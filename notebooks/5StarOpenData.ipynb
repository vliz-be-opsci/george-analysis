# 5-Star Open Data Model

The 5-star open data model describes a progression of data openness and usability. Each level builds on the previous one, improving the accessibility and interoperability of the data.  
(https://5stardata.info/en/)


#### ★ Data is Available on the Web
- **Definition:** Data is available online, in any format, under an open license.
- **Indicator:**
  - A publicly accessible URL where the data can be downloaded.
  - A clear, legally open license (e.g., Creative Commons, Open Data Commons).


#### ★★ Data is Structured and Machine-readable
- **Definition:** Data is in a structured format, making it easier for machines to process (e.g., CSV instead of PDF).
- **Indicator:**
  - File format is machine-readable (e.g., CSV, JSON, XML).
  - Content is clearly structured into rows, columns, or key-value pairs.

#### ★★★ Data Uses Non-proprietary Formats
- **Definition:** Data is available in an open, non-proprietary format.
- **Indicator:**
  - File format is open and documented (e.g., CSV, JSON, XML, instead of Excel or proprietary database formats).

#### ★★★★ Data Uses URIs for Identification
- **Definition:** Data items use URIs (Uniform Resource Identifiers) to uniquely identify resources, enabling linking.
- **Indicator:**
  - Dataset elements (e.g., rows, entities) use globally unique and resolvable URIs (e.g., `http://example.com/resource/123`).

#### ★★★★★ Data is Linked
- **Definition:** Data links to other datasets, providing context and enabling connections across datasets.
- **Indicator:**
  - Contains resolvable links to other datasets through RDF or linked data standards.
  - Data adheres to Linked Data principles, using standards like RDF, SPARQL, or JSON-LD.


### Summary Table

| **Star Level** | **Key Features**                         | **Verifiable Indicator**                                          |
|----------------|------------------------------------------|-------------------------------------------------------------------|
| ★ 1-Star       | Available online with open license       | Accessible URL and open license statement.                        |
| ★★ 2-Star      | Machine-readable format                  | File in CSV, JSON, or XML (or NetCDF) format.                     |
| ★★★ 3-Star     | Non-proprietary format                   | Data provided in open, non-proprietary formats (e.g., CSV).       |
| ★★★★ 4-Star    | URIs for identification                  | Unique and resolvable URIs for data elements.                     |
| ★★★★★ 5-Star   | Links to other datasets                  | Resolvable links to external data sources (RDF, SPARQL, JSON-LD). |

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract services and their star ratings
services = list(data.keys())
stars = list(data.values())

# Function to convert star ratings to emoji
def stars_to_emoji(value):
    full_stars = int(value)
    half_star = 0.5 if value - full_stars >= 0.5 else 0
    return "★" * full_stars + ("★️" if half_star else "")

# Set modern style
sns.set_theme(style="whitegrid")

# Plot the horizontal bar chart
plt.figure(figsize=(12, 8))
bars = plt.barh(services, stars, color=sns.color_palette("Blues", n_colors=len(services)), edgecolor='black')

# Add chart details
plt.xlabel('Stars (with penalization)', fontsize=14, labelpad=10)
plt.ylabel('Services', fontsize=14, labelpad=10)
plt.title('Star Ratings for Services', fontsize=18, pad=15)
plt.xlim(0, 5)
plt.gca().invert_yaxis()  # Invert for top-down order

# Annotate the star ratings on bars as stars
for bar, v in zip(bars, stars):
    plt.text(v + 0.1, bar.get_y() + bar.get_height() / 2, stars_to_emoji(v), 
             color='black', va='center', fontsize=12)

# Enhance layout and add grid
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
