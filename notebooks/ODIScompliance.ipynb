### ODIS framework

To assess compliance with the ODIS Framework, we used a scoring methodology to evaluate how effectively a service aligns with its key principles.  

Drawing on the ODIS Framework and its documentation (https://book.odis.org/), we defined 8 specific evaluation criteria, each scored on a predefined scale (e.g., 0-5), where higher scores reflect stronger alignment with the framework’s requirements.  

This systematic scoring process aims to provide a transparent and consistently applied indicator, designed to accommodate the diverse range of services under evaluation.

#### Identify Key ODIS Criteria

1. Data accessibility 
    - Data availability through standard web protocols (e.g., HTTP, APIs).
    - Open licensing and minimal restrictions on access.

2. Structured Data Publication
    - Machine-Readable Metadata: If the metadata is discoverable and readable by machines.
    - JSON-LD Implementation: Whether the service uses JSON-LD to provide structured metadata. (RDF dumps, json-ld within webpages, ...)
    - Schema.org Vocabulary Usage: Whether metadata is described through use of schema.org properties (& whether this is in aligment with ODIS thematic patterns).

3. Adherence to Thematic Patterns  
    - Adherence to ODIS thematic patterns ** for specific resource types (e.g., experts & institutions, documents (~ includes datasets), spatial maps, projects, training, vessels).  
    - Adherence to recognized metadata standards (e.g., ISO 19115, Dublin Core).  
    - Completeness of metadata (e.g., title, description, license, keywords).  

4. Interoperability
    - Use of controlled vocabularies (e.g., SeaDataNet, GCMD keywords).
    - Interoperable formats (e.g., NetCDF, XML, RDF).

5. Auto discovery 
    - Use of fair-signposting links
    - sitemaps protocol

6. Conformance declaration 
    - `conformsTo` statement from any URL (dataset, service, ...)

**sidenote: data offered by services also includes observations/measurements (~ single data points with fixed location & time), with are not explicitly mentioned within ODIS framework as a core profile.

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to plot radar chart for a single service
def plot_radar_chart(service_row):
    # List of criteria
    criteria = [
        "Data Accessibility", "Structured Data Publication", 
        "Adherence to Thematic Patterns", "Interoperability", 
        "Auto Discovery", "Conformance Declaration"
    ]

    # Get the service name and its scores
    service_name = service_row["Service"]
    scores = service_row[criteria].tolist()

    # Close the radar chart loop
    angles = np.linspace(0, 2 * np.pi, len(criteria), endpoint=False).tolist()
    angles += angles[:1]
    scores += scores[:1]

    # Create the radar chart
    fig, ax = plt.subplots(figsize=(8, 8), dpi=150, subplot_kw=dict(polar=True))
    ax.plot(angles, scores, linewidth=2, linestyle='solid', label=service_name)
    ax.fill(angles, scores, alpha=0.25)

    # Set up labels and title
    ax.set_yticklabels([])
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(criteria, fontweight="bold", fontsize=9)
    plt.title(f"Compliance for Service: {service_name}", size=14, weight='bold', pad=20)
    plt.legend(loc='upper right', fontsize=8)
    plt.tight_layout()
    plt.show()
```

```python
# Iterate through each service and plot
for _, row in df.iterrows():
    plot_radar_chart(row)
```
